```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split,GridSearchCV
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
x,y = load_diabetes(return_X_y=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape,y_train.shape)
```

```python
class SGDRegressor:
    def __init__(self,epochs=60,learning_rate=0.1):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.learning_rate = learning_rate
    def fit(self,x_train,y_train):
        x = x_train
        y = y_train
        self.intercept_ = 0
        self.coef_ = np.ones(x_train.shape[1])
        for i in range(self.epochs):
            for j in range(x_train.shape[0]):
                idx = np.random.randint(x_train.shape[0])
                y_hat = np.dot(x[idx],self.coef_)+self.intercept_
                slope_coe = (-2/x[idx].shape[0])*np.dot(y[idx]-y_hat,x[idx])
                slope_int = -2*np.mean(y[idx]-y_hat)
                self.intercept_ -= self.learning_rate*slope_int
                self.coef_ -= self.learning_rate*slope_coe
        return self.coef_,self.intercept_
    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
sgd = SGDRegressor()
sgd.fit(x_train,y_train)
y_pred = sgd.predict(x_test)
```

```python
n = x_train.shape[0]
k = x_train.shape[1]
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
adj_r2 = 1 - ((1-r2)*(n-1)/(n-1-k))
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R2 Score: {r2} ")
print(f"Adjusted R2 Score: {adj_r2}")
```
